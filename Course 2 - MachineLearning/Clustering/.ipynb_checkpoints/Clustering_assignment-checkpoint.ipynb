### Hopkins Score

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(countries_scaled)
```

Hopkins value indicates that although data is not highly clustered but it has clustering tendency and it is definetly not random.

## Modelling

### KMeans Clustering

### Optimal Number of Clusters

#### SSD

```python
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(countries_scaled)
    
    ssd.append(kmeans.inertia_)
    

# plot the SSDs for each n_clusters
plt.plot(ssd)
```

Note: here tick 0 represents 2 as value of K, 1 represents 3 and so on.

<b>As seen from ssd plot, 3 or 4 could be a good choice for K as after 4 there is no significant drop in inertia as K increases.</b>

#### Silhouette Analysis

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
silhouette_scores = []

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(countries_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(countries_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
    silhouette_scores.append(silhouette_avg)

plt.plot(silhouette_scores)
```

Note: here tick 0 represents 2 as value of K, 1 represents 3 and so on.

<b>clearly k=3 is good choice</b>

```python
# let's try another iteration of KMeans clustering with k=3
kmeans3 = KMeans(n_clusters=3, max_iter=50, random_state=100)
kmeans3.fit(countries_scaled)
```

```python
#Assigning labels
countries_3 = log_data.copy()
countries_3['country'] = labels
countries_3['label'] = kmeans3.labels_
countries_3.head()
```

In life_expec vs Child_mort we see a clear clustering done by KMeans

### Hierarchical Clustering

```python
countries_scaled.head()
```

```python
# single linkage
mergings = linkage(countries_scaled, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

Single Linkage is not very helpful in this case, trying with complete

```python
# complete linkage
mergings = linkage(countries_scaled, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

Again we'll choose 3 as a value for k, based on ssd and sillhuette Analysis. We can get three by cutting at 7 in dendrogram 

```python
# 3 clusters
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
#Assigning labels
countries_hc = log_data.copy()
countries_hc['country'] = labels
countries_hc['label'] = cluster_labels
countries_hc.head()
```

```python
sns.boxplot(x='label',y='income',data=countries_3)
```

Again, clusters are well seperated in terms of low and high income

```python
sns.scatterplot(data=countries_hc,x='child_mort',y='life_expec',hue='label',palette="deep")
```

As we see, countries with high life_expec and low child_mort are well seperated from countries under tough conditions  

```python
# Let's look at the pair plot with 3 desired variables 
sns.pairplot(data=countries_hc[['gdpp','child_mort','income','label']],diag_kind='kde',hue='label')
```

By comparing the Diagonals in pairplot og Hierarchical and KMeans clustering we see that Kmeans suggest lesser number of countries in direst need for ais i.e. cluster of countires with low income, low gdpp and high child_mort rate. 
Therefore I would prefer KMeans over Hierarchical as is we choose lesser number of countries (considering minimum 5) NGO would be able to help them the most
